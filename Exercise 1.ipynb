```python
#Reto Xaldigital Python
import pandas as pd #Utilizamos la librería pandas para el análisis de los datos
import requests
import json 
```

```python
#Conectarse al enlace
url = ('https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow')
req = requests.get(url) #Se hizo la petición para obtener el contenido de la url
obj_json = req.json() #Se inicializa la función obj_json
```

```python
#Inicializamos el dataframe con los valores descargados de la url 
df = pd.json_normalize(obj_json['items']) #Se normalizaron los datos json para tener un formato de tabla 
```

```python
#Desplegamos que se hayan cargado los valores correctamente 
df
```

```python
#Obtener el número de respuestas contestadas y no contestadas
total = df["is_answered"].count() #Total de respuestas
```

```python
#Respuestas contestadas 
answered = df[df["is_answered"] == True]["is_answered"].sum()
```

```python
#En las siguientes líneas de código desplegamos 
unanswered = total - answered
print("contestadas:", answered)
print("sin contestar:", unanswered)
```

```python
#Obtener la respuesta con menor número de vistas
view = df[ df ["view_count"] == df[ "view_count"].min()]
#Desplegamos el resultado de view para corroborar que obtenemos los valores esperados 
view
```

```python
#Obtener la respuesta más vieja y más actual 
old = df["creation_date"].min()

new = df["creation_date"].max()

print("más vieja:", old)
print("más actual:", new)
```

```python
#Obtener la respuesta del owner que tenga una mayor reputacion
reputation = df["owner.reputation"].max()
reputation
```
